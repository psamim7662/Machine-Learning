```python
# y = b0 + b1*x
def sigmoid(y):
    
    return (1 / (1+ np.exp(-y)))
    
```

```python
def p(x,b0,b1):
    return ( sigmoid(b0 + b1*x) )
```

```python
def cost(y,x,b0,b1):
    n = len(x)
    c = 0
    for i in range(n):
        c += y[i]*np.log(p(x[i],b0,b1) + 0.00001) + (1-y[i]) * np.log(1 - p(x[i],b0,b1)+0.0001)
    return -1*(c/n)
```
