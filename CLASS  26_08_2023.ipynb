# Histogram Equalization

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def showim(im1,im2):
    plt.subplot(121)
    plt.title('Original')
    plt.imshow(im1,cmap='gray')    
    plt.subplot(122)
    plt.title('Histogram equalized')
    plt.imshow(im2,cmap='gray')
    plt.show()
```

```python
img=cv2.imread("lena.tiff",0)
histeq=cv2.equalizeHist(img)
showim(img,histeq)
```

# Trackbar

```python
def nothing(x):
    print(x)
```

```python
cv2.namedWindow('image')
cv2.createTrackbar("test","image",50,500,nothing)
cv2.createTrackbar("color/gray","image",0,1,nothing)
```

```python
while (1):
    im=cv2.imread('lena.tiff')
    cv2.imshow('image',im)  
   
    test=cv2.getTrackbarPos("test","image")
    font=cv2.FONT_HERSHEY_COMPLEX
    cv2.putText(im,str(test),(50,150),font,4,(0,0,255))
    
    s=cv2.getTrackbarPos("color/gray","image")
    if s==0:
        pass
    else:
        im=cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
    cv2.imshow('image',im) 
    k=cv2.waitKey(1) & 0xFF
    if k==27 :
        break
    
cv2.destroyAllWindows()
```

# OBJECT DETECTION USING HSV COLOR SPACE

```python
import cv2
import numpy as np
```

```python
img=np.zeros((512,512,3),'uint8')
img[:,:,:]=(130,255,255)
cv2.imshow('color',img)
cv2.waitKey(0)
cv2.imshow('FRAME',im) 
cv2.destroyAllWindows()
```

```python
while (1):
    im=cv2.imread('smarties.jpg')
    hsv=cv2.cvtColor(im,cv2.COLOR_BGR2HSV)
    
    l_b=np.array([110,50,50])
    u_b=np.array([130,255,255])   
    
    mask=cv2.inRange(hsv,l_b,u_b)
    
    res=cv2.bitwise_and(im,im,mask=mask)
    
    cv2.imshow('FRAME',im) 
    cv2.imshow('HSV',hsv) 
    cv2.imshow('mask',mask) 
    cv2.imshow('res',res) 
    
    k=cv2.waitKey(0) & 0xFF
    if k==27 :
        break
    
cv2.destroyAllWindows()
```

# Object detection with trackbar

```python
import cv2
import numpy as np
```

```python
def operation(x):
     pass
```

```python
cv2.namedWindow('tracking')
cv2.createTrackbar("LH","tracking",0,255,operation)
cv2.createTrackbar("UH","tracking",0,255,operation)
cv2.createTrackbar("US","tracking",0,255,operation)
cv2.createTrackbar("LS","tracking",0,255,operation)
cv2.createTrackbar("UV","tracking",0,255,operation)
cv2.createTrackbar("LV","tracking",0,255,operation)
```

```python
while (1):
    im=cv2.imread('smarties.jpg') 
    hsv=cv2.cvtColor(im,cv2.COLOR_BGR2HSV)
    l_h=cv2.getTrackbarPos('LH','tracking')
    u_h=cv2.getTrackbarPos('UH','tracking')
    l_s=cv2.getTrackbarPos('LS','tracking')
    u_s=cv2.getTrackbarPos('US','tracking')  
    l_v=cv2.getTrackbarPos('LV','tracking')
    u_v=cv2.getTrackbarPos('UV','tracking')
    
    l_b=np.array([l_h,l_s,l_v])
    u_b=np.array([u_h,u_s,u_v]) 
    
    mask=cv2.inRange(hsv,l_b,u_b)
    
    res=cv2.bitwise_and(im,im,mask=mask)
    cv2.imshow('FRAME',im) 
    cv2.imshow('mask',mask) 
    cv2.imshow('res',res)     
    k=cv2.waitKey(1) & 0xFF
    if k==27 : 
        break
    
cv2.destroyAllWindows()
```
