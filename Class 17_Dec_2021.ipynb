# BIT WISE OPERATION

```python
import cv2
import numpy as np
```

```python
im1=cv2.imread('first.png',0)
im2=cv2.imread('second.png',0)
print(im1.shape)
print(im2.shape)
cv2.imshow('first',im1)
cv2.imshow('second',im2)
bit_and=cv2.bitwise_and(im2,im1)
cv2.imshow('and',bit_and)
bit_or=cv2.bitwise_or(im2,im1)
cv2.imshow('or',bit_or)
bit_xor=cv2.bitwise_xor(im2,im1)
cv2.imshow('xor',bit_xor)
bit_not=cv2.bitwise_not(im1)
cv2.imshow('not',bit_not)
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

# IMAGE BLENDING


```python
img1=cv2.imread('lena.tiff')
img2=cv2.imread('sailboat.tiff')
print(img1.shape)
print(img2.shape)
```

```python
img3=cv2.add(img1,img2)
img4=cv2.addWeighted(img1,0.3,img2,0.7,0)
```

```python
cv2.imshow('first',img1)
cv2.imshow('second',img2)
cv2.imshow('added',img3)
cv2.imshow('weighted',img4)
cv2.imwrite('lenablended.png',img4)
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

# method2

```python
img1=cv2.imread('lena.tiff')
img2=cv2.imread('splash.tiff')
```

```python
img2_gray=cv2.imread('splash.tiff',0)
```

```python
ret,mask=cv2.threshold(img2_gray,128,255,cv2.THRESH_BINARY)
mask_inv=cv2.bitwise_not(mask)
cv2.imshow('lena',img1)
cv2.imshow('splash',img2)
cv2.imshow('SPALSH GRAY',img2_gray)
cv2.imshow('mask ',mask)
cv2.imshow('inverse mask ',mask_inv)
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

```python
im1=cv2.bitwise_and(img1,img1,mask=mask)
im2=cv2.bitwise_and(img2,img2,mask=mask_inv)
result=cv2.add(im1,im2)
```

```python
cv2.imshow('lena',img1)
cv2.imshow('splash',img2)
cv2.imshow('mask ',mask)
cv2.imshow('inverse mask ',mask_inv)
cv2.imshow('lena background',im1)
cv2.imshow('splash no  background',im2)
cv2.imshow('Result ',result)
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

# Object detection with trackbar with Video

```python
import cv2
import numpy as np
```

```python
def nothing(x):
    print(x)
cap=cv2.VideoCapture("road_car_view.mp4")

cv2.namedWindow('tracking')
cv2.createTrackbar('LH','tracking',0,255,nothing)
cv2.createTrackbar('UH','tracking',255,255,nothing)
cv2.createTrackbar('LS','tracking',0,255,nothing)
cv2.createTrackbar('US','tracking',255,255,nothing)
cv2.createTrackbar('LV','tracking',0,255,nothing)
cv2.createTrackbar('UV','tracking',255,255,nothing)
```

```python
while (1):
    _,frame=cap.read()
    frame=cv2.resize(frame,(512,512),cv2.INTER_AREA)
    hsv=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    l_h=cv2.getTrackbarPos('LH','tracking')
    u_h=cv2.getTrackbarPos('UH','tracking')
    l_s=cv2.getTrackbarPos('LS','tracking')
    u_s=cv2.getTrackbarPos('US','tracking')  
    l_v=cv2.getTrackbarPos('LV','tracking')
    u_v=cv2.getTrackbarPos('UV','tracking')
    l_b=np.array([l_h,l_s,l_v])
    u_b=np.array([u_h,u_s,u_v])    
    mask=cv2.inRange(hsv,l_b,u_b)
    res=cv2.bitwise_and(frame,frame,mask=mask)
    cv2.imshow('FRAME',frame) 
    cv2.imshow('mask',mask) 
    cv2.imshow('res',res)     
    k=cv2.waitKey(1) & 0xFF
    if k==27 :
        break
cap.release()
cv2.destroyAllWindows()
```

# Thresholding

```python
import cv2
import numpy as np
```

```python
def nothing(x):
    print(x)
cv2.namedWindow('thresh')
cv2.createTrackbar('th','thresh',128,255,nothing)
```

```python
while True:
    im=cv2.imread('lena.tiff',0)
    t=cv2.getTrackbarPos('th','thresh')
    ret,binimage=cv2.threshold(im,t,255,cv2.THRESH_BINARY)
    cv2.imshow('thresh',im)
    cv2.imshow('Binary ',binimage)
    k=cv2.waitKey(1) & 0xFF
    if k==27 :
        break

cv2.destroyAllWindows()
```
