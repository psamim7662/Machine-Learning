```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
```

```python
im=cv2.imread("/lena.tiff",0)
plt.imshow(im,cmap='gray')
plt.show()
```

```python
pixel=im.flatten()
```

```python
pixel2=pixel.reshape(-1, 1)
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.cluster import KMeans  
wcss_list= []  #Initializing the list for the values of WCSS  
  
#Using for loop for iterations from 1 to 10.  
for i in range(1, 11):  
    kmeans = KMeans(n_clusters=i+2, init='k-means++', random_state= 42)  
    kmeans.fit(pixel2)  
    wcss_list.append(kmeans.inertia_)  
plt.plot(range(1, 11), wcss_list)  
plt.title('The Elobw Method Graph')  
plt.xlabel('Number of clusters(k)')  
plt.ylabel('wcss_list')  
plt.show()  
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state= 42)  
y_predict= kmeans.fit_predict(pixel2)  
```

```python
y_predict.shape
```

```python
c0=[]
c1=[]
c2=[]
c3=[]
c4=[]
for i in range(pixel.size):
  if y_predict[i]==0:
    c0.append(pixel[i])
  elif y_predict[i]==1:
    c1.append(pixel[i])
  elif y_predict[i]==2:
    c2.append(pixel[i])
  elif y_predict[i]==3:
    c3.append(pixel[i])
  elif y_predict[i]==4:
    c4.append(pixel[i])
```

```python
m0=np.mean(c0)
m1=np.mean(c1)
m2=np.mean(c2)
m3=np.mean(c3)
m4=np.mean(c4)
```

```python
newpixel=[]
for x in pixel:
  if x in c0:
    newpixel.append(m0)
  elif x in c1:
    newpixel.append(m1)
  elif x in c2:
    newpixel.append(m2)
  elif x in c3:
    newpixel.append(m3)
  elif x in c4:
    newpixel.append(m4)
```

```python
newimg=np.array(newpixel)
newim=newimg.reshape((im.shape[0],im.shape[1]))
```

```python
plt.imshow(newim,cmap='gray')
plt.show()
```
